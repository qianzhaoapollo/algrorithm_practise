```python
import matplotlib.pyplot as plt 
raw_data = []
after_filter = []
with open("out.txt", 'r') as f:
    for each in f:
        each = each.split()
        raw_data.append(float(each[0]))
        after_filter.append(float(each[1]))
length = len(raw_data)
x = range(length)
plt.plot(x, raw_data)
plt.plot(x, after_filter)
plt.show()
```
