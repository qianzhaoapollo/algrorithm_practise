```python
import pymysql
import pandas as pd
```

```python
import numpy as np
```

```python
db = pymysql.Connect("localhost", "root", "root", "testdb", charset='UTF8')
```

```python
cursor = db.cursor()
```

```python
sql = "SELECT substr(t.lzc, 6, LENGTH(t.lzc)-5) as stake from t_gd_congest t WHERE substr(t.insert_time, 1, 10) = '2018-02-21' \
AND substr(t.lzc, 1, 3) = 'G36' "
```

```python
cursor.execute(sql)
restult = cursor.fetchall()

res_dict = []
for field in cursor.description:
    res_dict.append(field[0])
```

```python
raw_res = pd.DataFrame(list(restult), columns=res_dict)
```
